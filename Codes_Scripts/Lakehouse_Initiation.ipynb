```python
%sql

-- create catelog for sales project 

CREATE CATALOG IF NOT EXISTS sales; 
USE CATALOG sales ; 


-- bronz layer schema creation 
--  this laye is used for landing raw data 
-- volume will be added on it to store physical raw files , then tables on top of it as archive for the raw data 

CREATE SCHEMA IF NOT EXISTS bronz_layer ;
USE bronz_layer; 
CREATE VOLUME IF NOT EXISTS sales.bronz_layer.raw_data;  -- use UI to upload data on this VLOUME


-- silver layer schema creation 
-- this schema is the single source of truth as the data is cleaned 

CREATE SCHEMA IF NOT EXISTS sales.silver_layer;



-- gold layer schema creation 
-- final layer of the Medallion Architecture, this schema contains Consumption-Ready data 
-- data is aggregated, fileterd and joined for bussiness users and data applications 
CREATE SCHEMA IF NOT EXISTS sales.gold_layer; 
```
